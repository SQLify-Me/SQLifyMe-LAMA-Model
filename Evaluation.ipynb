```python
from langchain_community.embeddings.huggingface import HuggingFaceEmbeddings
embedding_model = HuggingFaceEmbeddings()
```

```python
from langchain.evaluation import load_evaluator

hf_evaluator = load_evaluator("embedding_distance", embeddings=embedding_model)
```

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
eval_data=pd.read_excel("/content/Llama2_eval.xlsx")
```

```python
# Function to evaluate the distance between two strings
def evaluate_strings(prediction, reference):
  eval=hf_evaluator.evaluate_strings(prediction=prediction, reference=reference)
  return eval['score']

# Apply the function to the DataFrame
eval_data['distance'] = eval_data.apply(lambda row: evaluate_strings(row['sql'], row['generatedSQL']), axis=1)

# Print the DataFrame with the calculated distances
print(eval_data)
```

```python
eval_data.head()
```

```python
# Calculate the average distance
average_distance = eval_data['distance'].mean()
```

```python
print("Llama 2  aveage distance: ",average_distance)
```
